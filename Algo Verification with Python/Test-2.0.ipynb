```python
def mean_normal(a,r):
    temp = np.zeros(a.shape, dtype = float)
    yMean = np.zeros((1682, 1), dtype=float)
    for i in range(len(r)):
        mean = sum(a[i]) / sum(r[i])
        
        temp[i] = a[i] - mean
        yMean[i]=mean
       
    return temp * r, yMean;
```

```python
alpha = 0.001
Lambda = .001
num_iters = 60
n = 25
```

```python
movies = 1682
user = 943
X = np.random.rand(movies,n)
Theta = np.random.rand(user,n)
X
```

```python
jlist = list()
iterlist = list()
for i in range(1, num_iters):
    Error = ((X.dot(np.transpose(Theta)))-yPrime).copy()
    X_grad = ((Error*r).dot(Theta) + Lambda * X).copy()
    Theta_grad = ((np.transpose(Error*r)).dot(X) + Lambda * Theta).copy()
    X = (X - (alpha * X_grad)).copy()
    Theta = (Theta - (alpha * Theta_grad)).copy()
    J = (.5) * ((Error * Error)* r).sum()
    Reg_term_theta = (Lambda / 2) * ((Theta * Theta).sum())
    Reg_term_X = (Lambda / 2) * ((X * X).sum())
    J = J + Reg_term_theta + Reg_term_X
    jlist.append(J)
    iterlist.append(i)

print(jlist)
print(iterlist)
```

```python
plt.plot(iterlist, jlist, color= 'b', marker = 'o')

plt.xlabel("Iterations")
plt.ylabel("cost Function, J")
plt.title("Alpha = 0.001, Lambda = 0.001, numIter = 60, Feature = 25")
plt.show()
```
