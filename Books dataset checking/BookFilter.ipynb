```python
import pandas as pd
```

```python
df = pd.read_csv("books.csv", error_bad_lines=False)
```

```python
df.head()
```

```python
df.rename(columns={"bookID": "book_id", "title": "name"}, inplace=True)
df.head(20)
```

```python
filt = (df["average_rating"] >= 4.0)
filtered_book = df[["book_id", "name", 
    "authors", "ratings_count", 
    "text_reviews_count", 
    "average_rating"]][filt].sort_values(["ratings_count","text_reviews_count"], ascending=False)[["book_id", 
                                                                                                  "name", 
                                                                                                  "authors"]][:1682]
filtered_book
```
